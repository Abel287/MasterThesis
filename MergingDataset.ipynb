```python
import pandas as pd
import xlrd
from xlrd import open_workbook
import networkx as nx
import matplotlib.pyplot as plt
import warnings
import scipy as sp
import pprint
import operator
from math import sin, cos, sqrt, atan2, radians
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
excel_file ='Datasets Combined.xlsx'
workbook= pd.read_excel(excel_file, sheet_name='Sheet1',index_col=0)
node_count = workbook.shape[0]
book = open_workbook('Datasets Combined.xlsx')
node_data = book.sheet_by_name('Sheet1') #read node file and correct sheet
NodeLabel={} #create dictionary to name nodes based on station label
NodeCodes={}
for i in range(node_count):
    NodeLabel[int(node_data.cell_value(i+1,0))]=node_data.cell_value(i+1,1)
    NodeCodes[int(node_data.cell_value(i+1,0))]=[]
    for j in range(2,len(workbook.columns)):
        if node_data.cell_value(i+1,j) != '':
            NodeCodes[int(node_data.cell_value(i+1,0))].append(int(node_data.cell_value(i+1,j)))
```

```python
OD_file='S2S.xlsx'
OD_data_sheet= pd.read_excel(OD_file, sheet_name='Sheet1',index_col=0)
OD_data_count = OD_data_sheet.shape[0]
OD_data_book=open_workbook('S2S.xlsx')
OD_data = OD_data_book.sheet_by_name('Sheet1') #read node file and correct sheet
```

```python
Origin=[]
Destination=[]
for i in range(OD_data_count):
    Origin.append(int(OD_data.cell_value(i+1,0)))
    Destination.append(int(OD_data.cell_value(i+1,0)))
Demand={}
for i in range(1,OD_data_count+1):
    Demand[int(OD_data.cell_value(i,0))]={}
    for j in range(1,OD_data_count+1):
        Demand[int(OD_data.cell_value(i,0))][int(OD_data.cell_value(0,j))]=OD_data.cell_value(i,j)
```

```python
Dim={}
for i in NodeCodes:
    Dim[i]={}
    for j in range(1,OD_data_count+1):
        Dim[i][int(OD_data.cell_value(0,j))]=0
        for k in NodeCodes[i]:
            Dim[i][int(OD_data.cell_value(0,j))]+= Demand[k][int(OD_data.cell_value(0,j))]
D={}
for i in NodeCodes:
    D[i]={}
    for j in NodeCodes:
        D[i][j]=0
        if i != j:
            for k in NodeCodes[j]:
                D[i][j]+=Dim[i][k]
        
```

```python
import xlsxwriter

# Create a new Excel file and add a worksheet.
workbook = xlsxwriter.Workbook('DemandData.xlsx')
worksheet = workbook.add_worksheet()
worksheet.write(0,0,"Origin/Destination")
k=1
for i in D:
    worksheet.write(k, 0, i)
    
    l=1
    for j in D[i]:
        worksheet.write(0, l, j)
        worksheet.write(k, l, D[i][j])
        l += 1
    k += 1
workbook.close()
```
